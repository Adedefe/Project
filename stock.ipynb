```python
train_ind = int(len(data)*0.8)
train = data[:train_ind]
test = data[train_ind:]
```

```python
from pmdarima.arima import auto_arima


model = auto_arima(train['Visa'], start_p=1, start_q=1,
                      test='adf',
                      max_p=5, max_q=5,
                      m=1,             
                      d=1,          
                      seasonal=False,   
                      start_P=0, 
                      D=None, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
```

```python
model.summary()
```

```python
# Forecast

prediction, confint = model.predict(n_periods=len(test['Visa']), return_conf_int=True)

prediction
```

```python
cf= pd.DataFrame(confint)
```

```python
prediction_series = pd.Series(prediction,index=test.index)
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
ax.plot(data['Visa'])
ax.plot(prediction_series)
ax.fill_between(prediction_series.index,
                cf[0],
                cf[1],color='grey',alpha=.3)
```

```python
model.plot_diagnostics(figsize=(14,10))
plt.show()
```
