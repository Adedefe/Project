```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating random square feet and House prices with equal length

x=[45,58,42,66,74,60,83.7,50,90.5]
y=[49500,68950,47000,71000,75890,69900,94300,62000,101540]
len(x)==len(y)
```

```python
#converting into array of lists

x=np.array(x)
y=np.array(y)
```

```python
#Scatter plot of input versus output
plt.scatter(x,y)
plt.xlabel('Square ft')
plt.ylabel('Price')
plt.title('House price scatter plot')
plt.grid()
plt.show()
```

# 

```python
# calculating for the Yi-Ŷ, then putting into list
z=[]
for i in y:
    z.append((i-(sum(y)/len(y))))
```

```python
# calculating for the Xi-x(hat), then putting into list
w=[]
for i in x:
    w.append((i-(sum(x)/len(x))))
```

```python
# summing product of z and w arrays
t=sum(np.array(z)*np.array(w))
t
```

```python
# taking second power of product of w array, then summing
r=sum(np.array(w)**2)
r
```

```python
slope=t/r
slope
```

```python
intercept=y.mean()-(slope*(x.mean()))
intercept
```

```python
y_pred=intercept+slope*x
```

```python
# predicting new House:
x_new=73.7
y_pred=intercept+slope*x_new
print('Approximate price for the new house:',y_pred,'U.S dollars')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y,  color='red')
plt.plot(x, y_pred, color='blue', linewidth=3)
plt.grid()
```
